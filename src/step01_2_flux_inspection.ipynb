```python
import pandas as pd


control = pd.read_parquet("results/fluxes/flux_samples_CONTROL_10_000.parquet.gzip")
pku     = pd.read_parquet("results/fluxes/flux_samples_PKU_10_000.parquet.gzip")
```

```python

control['Group'] = 'control'
pku['Group'] = 'pku'

samples = pd.concat([control, pku], axis=0).reset_index(drop=True)
```

```python

import seaborn as sns






sns.swarmplot(data=samples.sample(500)[['DHPR', 'Group']], x="Group", y="DHPR")


#sns.swarmplot(data=samples.sample(500)[['DHPR', 'Group']], x="Group", y="DHPR")
```
