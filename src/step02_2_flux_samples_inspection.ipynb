```python
import pandas as pd


control = pd.read_parquet("./results/fluxes/flux_samples_CONTROL_20_000.parquet.gzip")
pku     = pd.read_parquet("./results/fluxes/flux_samples_PKU_20_000.parquet.gzip")
```

```python

control['Group'] = 'control'
pku['Group'] = 'pku'

samples = pd.concat([control, pku], axis=0).reset_index(drop=True)
import seaborn as sns

sns.swarmplot(data=samples.sample(500)[['r0399', 'Group']], x="Group", y="r0399")
```

```python
sns.swarmplot(data=samples.sample(2000)[['C30CPT1', 'Group']], x="Group", y="C30CPT1")
```
